```python
import pandas as pd
```

```python
LEXER_TABLE = pd.DataFrame(columns=['TOKEN','IDENTITY'],index=range(1))
```

```python
def lexer():
    with open('ip.txt', 'r') as myfile:
        string = myfile.read().replace('\n', '')
    print(string)
    SYMBOLS = ['(', ')', ';', ',']
    SYMBOLS_1 = ['(', ')', ';', ',', '+', '<=', '++', '=']
    SYMBOLS_1_DUP = [['Open Bracket', '('], ['Close Bracket', ')'], ['Semicolon', ';'], ['Comma', ','], [
        'Plus', '+'], ['Lesser than or equal to', '<='], ['Increment', '++'], ['Equals', '=']]
    keywords = ['int', 'main', 'begin', 'for', 'end', ' ', '\n']
    keywords_def = [['t', 'int'], ['m', 'main'], ['b', 'begin'], ['f', 'for'], [
        'd', 'end'], ['s', ' '], ['n', '\n'], ['o', '+'], ['o', '<='], ['o', '++'], ['o', '=']]
    KEYWORDS = SYMBOLS_1+keywords
    white_space = ' '
    lexeme = ''
    list = []
    string = string.replace('\t', '')
    skip = False
    for i, char in enumerate(string):
        if skip == True:
            skip = False
            continue
        if char != white_space:
            lexeme += char
        if(i+1 < len(string)):
            if string[i+1] == white_space or string[i+1] in KEYWORDS or lexeme in KEYWORDS:
                if lexeme != '':
                    if lexeme == '+' and string[i+1] == '+':
                        list.append('++')
                        lexeme = ''
                        skip = True
                        continue
                    elif lexeme == '<' and string[i+1] == '=':
                        list.append('<=')
                        lexeme = ''
                        skip = True
                        continue
                    list.append(lexeme.replace('\n', '<newline>'))
                    lexeme = ''
    list.append(lexeme.replace('\n', '<newline>'))
    # print(list)
    s = ''
    j = 0
    for item in list:
        for i in keywords_def:
            if i[1] == item:
                s = s+i[0]
        if item in SYMBOLS:
            s = s+item
        if item.isdigit() or item not in KEYWORDS:
            s = s+'v'

    flag = False
    for i in list:
        flag = False
        for k in SYMBOLS_1_DUP:
            if i == k[1]:
                LEXER_TABLE.at[j, 'TOKEN'] = i
                LEXER_TABLE.at[j, 'IDENTITY'] = k[0]
                j += 1
                flag = True
                break
        if flag == False:
            if i in keywords:
                LEXER_TABLE.at[j, 'TOKEN'] = i
                LEXER_TABLE.at[j, 'IDENTITY'] = 'Keyword'
                j += 1
                continue
            if i.isdigit():
                LEXER_TABLE.at[j, 'TOKEN'] = i
                LEXER_TABLE.at[j, 'IDENTITY'] = 'Digit'
                j += 1
                continue
            elif i not in keywords:
                LEXER_TABLE.at[j, 'TOKEN'] = i
                LEXER_TABLE.at[j, 'IDENTITY'] = 'Identifier'
                j += 1
    return s
```

```python
lex_str = lexer()
lex_str
```
